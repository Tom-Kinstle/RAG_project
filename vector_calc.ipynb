Given 3D vectors [40,120,60] and [60,50,90]

### Q1 draw out these vectors

```python
import numpy as np
import matplotlib.pyplot as plt

#Initialize the two vectors

colour1 = [40, 120, 60]
colour2 = [60, 50, 90]

ax = plt.figure().add_subplot(projection='3d')

# Draw out the vectors
ax.scatter(xs=colour1[0], ys=colour1[1], zs=colour1[2], zdir='z', label='Colour 1')
ax.scatter(xs=colour2[0], ys=colour2[1], zs=colour2[2], zdir='z', label='Colour 2')

# origin [0,0,0] is the default for scatter, so we don't need to specify it
ax.scatter( 0, 0, 0, zdir='z', label='Origin', c='black')

ax.quiver(0, 0, 0, colour1[0], colour1[1], colour1[2], color='black', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, colour2[0], colour2[1], colour2[2], color='black', arrow_length_ratio=0.1)

# Make legend, set axes limits and labels
ax.legend()
ax.set_xlim(0, 150)
ax.set_ylim(0, 150)
ax.set_zlim(0, 150)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Customize the view angle so it's easier to see that the scatter points lie
# on the plane y=0
ax.view_init(elev=20., azim=-35, roll=0)

plt.show()
```

### Q2 Calculate the euclidian, manhattan, cosine, and dot product distances between these vectors

### Manhattan distance

```python
L1 = [colour1[i] - colour2[i] for i in range(len(colour1))]
L1 = np.abs(L1).sum()
print(L1)
```

```python
#An alternate approach

a = np.array(colour1)
b = np.array(colour2)

np.linalg.norm((a - b), ord=1)
```

### Euclidian distance

```python
L2 = [(colour1[i] - colour2[i]) **2 for i in range(len(colour1))]
L2 = np.sqrt(np.sum(L2))
print(L2)
```

```python
np.linalg.norm((a - b), ord=2)
```

### Cosine distance

```python
cosine = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
print(cosine)
```

### Dot product

```python
dot = np.array([colour1[i] * colour2[i] for i in range(len(colour1))]).sum()
print(dot)
```

```python
#alternate method

np.dot(a,b)
```
